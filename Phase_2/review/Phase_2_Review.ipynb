```python
# Null: Foreign films perform similarly at the box office when compared to non-foreign films.
# Alternative: Foreign films perform differently at the box office when compared to non-foreign films.

foreign_films = df[df['country'] != 'USA']['gross'].dropna().values
domestic_films = df[df['country'] == 'USA']['gross'].dropna().values
stats.ttest_ind(foreign_films, domestic_films)

# Foreign films do perform differently at the box offic when compared to domestic films.
# The pvalue of 3.86e-33 is smaller than our alpha of .05 which means we reject the null hypothesis.
```

Explain what a Type I error is and how it relates to the significance level when doing a statistical test. 

```python
# A type 1 error is a false positive and occurs when we reject the null hypothesis even though we should have
# failed to reject the null hypothesis. By random chance we may end up finding an effect where one doesn't exist
# We expect these when we perform many tests when there is no effect.
# The significance level, alpha, is the level we set, generally between 1-5%, that defines
# our tolerance for a type 1 error. If our p value, or probability of the statistical test
# being true, is less than that error tolerance we reject the null hypothesis because
# the chance of it occuring is so low it is likely false.

# so if our significance level is 95%, what does type 1 error mean.
    # means 5% chance of making a Type 1 Error
```
